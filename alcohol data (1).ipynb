```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Alcohol data set. This shows average alcohol consumption per capita for those age 15 and over in each country.

df = pd.read_csv("/Users/elijahwoolford/Desktop/python /drinks.csv")
df
```

```python
df2 = pd.read_csv("/Users/elijahwoolford/Desktop/python /countries.csv")
#df2
```

```python
dic = {}
country = []
continent = []
beer = []
wine = []
for i in range(len(df["country"])):
    for j in range(len(df2["name"])):
        if df["country"][i] == df2["name"][j]:
            dic[i] = df2["region"][j]
dic[184] = "Americas"
dic[182] = "Europe"
for key in dic:
    country.append(df.loc[key, "country"])
    continent.append(dic[key])
    beer.append(df.loc[key, "beer_servings"])
    wine.append(df.loc[key, "wine_servings"])
          
df3 = pd.DataFrame()
df3["country"] = country
df3["continent"] = continent
df3["beer"] = beer
df3["wine"] = wine
```

```python
average = df3.groupby("continent").mean()
average
```

```python
# Two scatterplots showing correlation between type of alcohol and total alcohol consumed. This shows that countries 
# tend to drink more beer overall than wine.

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.scatter(x=df["beer_servings"], 
            y=df["total_litres_of_pure_alcohol"],
            color="purple")
plt.xlabel("beer servings")
plt.ylabel("total litres of pure alcohol")
plt.title("Total beer vs Total liters of alcohol")

plt.subplot(1,2,2)
plt.scatter(x=df["wine_servings"],
            y=df["total_litres_of_pure_alcohol"],
            color="red")
plt.xlabel("wine servings")
plt.ylabel("total litres of pure alcohol")
plt.title("Total wine vs Total liters of alcohol")

plt.tight_layout()
plt.show()
```

```python
# Two boxplots that show the ranges of beer and wine servings. This backs up the claim that people drink more
# beer than alcohol in the world, with the average number of beer servings being more than average wine servings

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.boxplot(df["beer_servings"])

plt.subplot(1,2,2)
plt.boxplot(df["wine_servings"])

plt.show()
```

```python
#Two plots that show average beer and wine per continent.

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(average["beer"])
plt.xlabel("Continent")
plt.ylabel("Average Beer")

plt.subplot(1,2,2)
plt.plot(average["wine"])
plt.xlabel("Continent")
plt.ylabel("Average Wine")

plt.show()
```
